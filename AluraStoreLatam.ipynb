#1. Análisis de facturación



```python
# Cargar los datos
tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

```python
# Calcular ingresos totales para cada tienda
ingreso_tienda1 = tienda1['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()
```

```python
# Mostrar los resultados
print(f"Ingreso total Tienda 1: {ingreso_tienda1}")
print(f"Ingreso total Tienda 2: {ingreso_tienda2}")
print(f"Ingreso total Tienda 3: {ingreso_tienda3}")
print(f"Ingreso total Tienda 4: {ingreso_tienda4}")
```

Con los resultados exuestos anteriormente, podemos definir que la tienda 1 tiene mayores ingresos y la tienda 4 tiene menores ingresos

# 2. Ventas por categoría

```python
tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

```python
# Función para analizar una tienda
def analizar_tienda(df, nombre_tienda):
    print(f"\nAnálisis para {nombre_tienda}:")
    conteo = df['Categoría del Producto'].value_counts().sort_values(ascending=False)
    print(conteo)
    producto_mas_vendido = conteo.idxmax()
    print(f"\nEl producto más vendido en {nombre_tienda} es: {producto_mas_vendido} con {conteo.max()} ventas")
```

```python
# Analizar cada tienda
analizar_tienda(tienda1, "Tienda 1")
analizar_tienda(tienda2, "Tienda 2")
analizar_tienda(tienda3, "Tienda 3")
analizar_tienda(tienda4, "Tienda 4")
```

# 3. Calificación promedio de la tienda


```python
# Cargar los datos
tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Calcular promedios
promedio_tienda1 = tienda1['Calificación'].mean()
promedio_tienda2 = tienda2['Calificación'].mean()
promedio_tienda3 = tienda3['Calificación'].mean()
promedio_tienda4 = tienda4['Calificación'].mean()

# Mostrar resultados
print(f"Promedio de calificación Tienda 1: {promedio_tienda1:.2f}")
print(f"Promedio de calificación Tienda 2: {promedio_tienda2:.2f}")
print(f"Promedio de calificación Tienda 3: {promedio_tienda3:.2f}")
print(f"Promedio de calificación Tienda 4: {promedio_tienda4:.2f}")
```

En promedio, la tienda mejor valorada por los clientes, es la tienda 3

# 4. Productos más y menos vendidos

```python
# Leer los datos
tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Función para analizar una tienda
def analizar_tienda(df, nombre_tienda):
    print(f"\nAnálisis para {nombre_tienda}:")

    # Contar ventas por producto
    ventas_por_producto = df['Producto'].value_counts()

    # Productos más vendidos (top 3)
    print("\nProductos más vendidos:")
    print(ventas_por_producto.head(3))

    # Productos menos vendidos (bottom 3)
    print("\nProductos menos vendidos:")
    print(ventas_por_producto.tail(3))

# Analizar cada tienda
analizar_tienda(tienda1, "Tienda 1")
analizar_tienda(tienda2, "Tienda 2")
analizar_tienda(tienda3, "Tienda 3")
analizar_tienda(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
# Calcular promedios
promedio_tienda1 = tienda1["Costo de envío"].mean()
promedio_tienda2 = tienda2["Costo de envío"].mean()
promedio_tienda3 = tienda3["Costo de envío"].mean()
promedio_tienda4 = tienda4["Costo de envío"].mean()

# Mostrar resultados
print(f"Promedio costo de envío Tienda 1: {promedio_tienda1:.2f}")
print(f"Promedio costo de envío Tienda 2: {promedio_tienda2:.2f}")
print(f"Promedio costo de envío Tienda 3: {promedio_tienda3:.2f}")
print(f"Promedio costo de envío Tienda 4: {promedio_tienda4:.2f}")
```

Analisis en  graficos

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos proporcionados
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [1150880400.0, 1116343500.0, 1098019600.0, 1038375700.0]
calificaciones = [3.98, 4.04, 4.05, 4.00]
costo_envio = [26018.61, 25216.24, 24805.68, 23459.46]

# Datos de categorías (usaremos Tienda 1 como ejemplo)
categorias = ['Muebles', 'Electrónicos', 'Juguetes', 'Electrodomésticos',
              'Deportes', 'Instrumentos', 'Libros', 'Hogar']
ventas_tienda1 = [465, 448, 324, 312, 284, 182, 173, 171]

# 1. Gráfico de barras - Ingresos por tienda
plt.figure(figsize=(10, 6))
bars = plt.bar(tiendas, ingresos, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Ingresos Totales por Tienda', fontsize=16)
plt.ylabel('Ingresos (en miles de millones)', fontsize=12)
plt.xticks(fontsize=10)

# Formatear los valores en el eje Y
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:.1f}B".format(x/1e9)))

# Añadir etiquetas con los valores exactos
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             '{:,.0f}'.format(height),
             ha='center', va='bottom')

plt.tight_layout()
plt.show()

# 2. Gráfico circular - Distribución de ventas por categoría (Tienda 1)
plt.figure(figsize=(8, 8))
explode = (0.1, 0, 0, 0, 0, 0, 0, 0)  # Destacar la categoría más vendida
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6','#c4e17f','#ffd700']

plt.pie(ventas_tienda1, explode=explode, labels=categorias, colors=colors,
        autopct='%1.1f%%', startangle=90, pctdistance=0.85, textprops={'fontsize': 10})
plt.title('Distribución de Ventas por Categoría - Tienda 1', fontsize=14)

# Dibujar un círculo en el centro para que sea un donut (opcional)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.tight_layout()
plt.show()

# 3. Gráfico combinado - Calificaciones y costos de envío
fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:blue'
ax1.set_xlabel('Tienda')
ax1.set_ylabel('Calificación Promedio', color=color)
ax1.plot(tiendas, calificaciones, color=color, marker='o', linestyle='--', linewidth=2, markersize=8)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(3.9, 4.1)  # Rango para mejor visualización

# Instanciar un segundo eje que comparte el mismo eje x
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Costo Promedio de Envío', color=color)
ax2.bar(tiendas, costo_envio, color=color, alpha=0.4)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Calificación Promedio y Costo de Envío por Tienda', fontsize=16)
fig.tight_layout()

# Añadir etiquetas de valores
for i, v in enumerate(calificaciones):
    ax1.text(i, v+0.01, str(v), color='blue', ha='center')

for i, v in enumerate(costo_envio):
    ax2.text(i, v+100, f"{v:,.0f}", color='red', ha='center')

plt.show()
```
